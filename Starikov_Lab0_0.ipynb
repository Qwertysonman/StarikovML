```python
data = pd.read_csv('data.csv')
data
```

```python
fig1, ax1 = plt.subplots()
counts = list(data.groupby('Rooms').count()['Id'])
name = data['Rooms'].unique()
explode = np.zeros(len(counts))
explode[counts.index(max(counts))] = 0.2
ax1.pie(counts, labels = name, autopct='%1.1f%%', explode = explode)
plt.legend(counts, loc = 'upper left')
plt.show()
```

```python
price_data = list(data['Price'])
plt.hist(price_data, 100, edgecolor = 'black')
plt.grid()
plt.show()
```

```python

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Построение диаграмм рассеяния для каждого признака
data.plot.scatter(x='Rooms', y='Price', ax=axes[0, 0])
data.plot.scatter(x='Square', y='Price', ax=axes[0, 1])
data.plot.scatter(x='HouseFloor', y='Price', ax=axes[1, 0])
data.plot.scatter(x='HouseYear', y='Price', ax=axes[1, 1])

# Настройка заголовков для каждой диаграммы
axes[0, 0].set_title('Rooms vs Price')
axes[0, 1].set_title('Square vs Price')
axes[1, 0].set_title('HouseFloor vs Price')
axes[1, 1].set_title('HouseYear vs Price')

# Отображение диаграмм рассеяния
plt.tight_layout()
plt.show()
```
